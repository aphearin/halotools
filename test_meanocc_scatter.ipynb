```python
%matplotlib inline 
from astropy.table import Table
from halotools.utils.table_utils import compute_conditional_percentiles
from halotools.utils.array_utils import randomly_downsample_data

from matplotlib import lines as mlines

Npts = 1e4
mass = np.logspace(11.5, 13, Npts)
conc = np.random.random(Npts)
d = {'halo_mvir': mass, 'halo_nfw_conc': conc}
halo_table = Table(d)
halo_table['halo_nfw_conc_percentile'] = compute_conditional_percentiles(halo_table=halo_table, 
                                                   prim_haloprop_key='halo_mvir', 
                                                   sec_haloprop_key='halo_nfw_conc', 
                                                   dlog10_prim_haloprop=0.05)
halo_table.sort('halo_mvir')

highz_mask = halo_table['halo_nfw_conc_percentile'] >= 0.5
highz_halos = halo_table[highz_mask]
lowz_halos = halo_table[np.invert(highz_mask)]

nhalos = np.min([len(highz_halos), len(lowz_halos)])
highz_halos = randomly_downsample_data(highz_halos, nhalos)
lowz_halos = randomly_downsample_data(lowz_halos, nhalos)

highz_halos.sort('halo_mvir')
lowz_halos.sort('halo_mvir')
```

```python
from halotools.empirical_models import Hearin15
baseline_model20 = Hearin15(threshold=10, satellite_assembias=False, central_assembias=False)
baseline_model80 = Hearin15(threshold=10, satellite_assembias=False, central_assembias=False)
baseline_model80.param_dict['scatter_model_param1_centrals'] = 0.8

assembias_model20 = Hearin15(threshold=10)
assembias_model80 = Hearin15(threshold=10)
assembias_model80.param_dict['scatter_model_param1_centrals'] = 0.8
```

```python
lowz_ncen_baseline20 = baseline_model20.mean_occupation_centrals(halo_table=lowz_halos)
highz_ncen_baseline20 = baseline_model20.mean_occupation_centrals(halo_table=highz_halos)
lowz_ncen_baseline80 = baseline_model80.mean_occupation_centrals(halo_table=lowz_halos)
highz_ncen_baseline80 = baseline_model80.mean_occupation_centrals(halo_table=highz_halos)

lowz_ncen_assembias20 = assembias_model20.mean_occupation_centrals(halo_table=lowz_halos)
highz_ncen_assembias20 = assembias_model20.mean_occupation_centrals(halo_table=highz_halos)
lowz_ncen_assembias80 = assembias_model80.mean_occupation_centrals(halo_table=lowz_halos)
highz_ncen_assembias80 = assembias_model80.mean_occupation_centrals(halo_table=highz_halos)
```
